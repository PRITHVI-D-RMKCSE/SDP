```python
class Solution:
    def solve(self, board):
        # Dictionary to track visited states
        state_dict = {}
        # Flatten the board into a tuple for easier manipulation
        flatten = [num for row in board for num in row]
        flatten = tuple(flatten)
        state_dict[flatten] = 0

        # Check if the initial board is already the solution
        if flatten == (0, 1, 2, 3, 4, 5, 6, 7, 8):
            return 0

        return self.get_paths(state_dict)

    def get_paths(self, state_dict):
        cnt = 0
        while True:
            current_nodes = [x for x in state_dict if state_dict[x] == cnt]

            # If there are no current nodes, return -1 (unsolvable state)
            if not current_nodes:
                return -1

            for node in current_nodes:
                next_moves = self.find_next(node)
                for move in next_moves:
                    if move not in state_dict:
                        state_dict[move] = cnt + 1
                        # Check if the solution is found
                        if move == (0, 1, 2, 3, 4, 5, 6, 7, 8):
                            return cnt + 1
            cnt += 1

    def find_next(self, node):
        # Possible moves for each index
        moves = {
            0: [1, 3],
            1: [0, 2, 4],
            2: [1, 5],
            3: [0, 4, 6],
            4: [1, 3, 5, 7],
            5: [2, 4, 8],
            6: [3, 7],
            7: [4, 6, 8],
            8: [5, 7],
        }

        results = []
        pos_0 = node.index(0)
        for move in moves[pos_0]:
            new_node = list(node)
            # Swap positions to generate the next state
            new_node[move], new_node[pos_0] = new_node[pos_0], new_node[move]
            results.append(tuple(new_node))

        return results

# Example usage
if __name__ == "__main__":
    ob = Solution()
    matrix = [
        [3, 1, 2],
        [4, 7, 5],
        [6, 8, 0]
    ]
    print("Minimum moves to solve the puzzle:", ob.solve(matrix))
```

```python
print("Enter the number of queens")
N = int(input())

# Create a chessboard NxN matrix with all elements set to 0
board = [[0] * N for _ in range(N)]

def is_safe(i, j):
    # Check vertically and horizontally
    for k in range(N):
        if board[i][k] == 1 or board[k][j] == 1:
            return False

    # Check diagonally
    for k in range(N):
        for l in range(N):
            if (k + l == i + j or k - l == i - j) and board[k][l] == 1:
                return False

    return True

def solve_n_queens(n):
    if n == 0:
        return True

    for i in range(N):
        for j in range(N):
            if not is_safe(i, j):
                continue
            if board[i][j] != 1:
                board[i][j] = 1
                if solve_n_queens(n - 1):
                    return True
                board[i][j] = 0

    return False

# Check if a solution exists
if solve_n_queens(N):
    print("Solution exists. Placements of queens:")
    for row in board:
        print(row)
else:
    print("No solution exists.")
```

```python
from collections import defaultdict

class Graph:
    def __init__(self):
        # Default dictionary to store the graph
        self.graph = defaultdict(list)

    # Function to add an edge to the graph
    def add_edge(self, u, v):
        self.graph[u].append(v)

    # A utility function for the DFS algorithm
    def _dfs_util(self, v, visited):
        # Mark the current node as visited and print it
        visited.add(v)
        print(v, end=' ')

        # Recur for all adjacent vertices
        for neighbour in self.graph[v]:
            if neighbour not in visited:
                self._dfs_util(neighbour, visited)

    # Function to perform DFS traversal
    def dfs(self, v):
        # Create a set to store visited vertices
        visited = set()
        # Call the recursive helper function to print DFS traversal
        self._dfs_util(v, visited)

    # Function to perform BFS traversal
    def bfs(self, s):
        # Create a queue for BFS
        queue = []
        visited = []

        # Mark the source node as visited and enqueue it
        queue.append(s)
        visited.append(s)

        while queue:
            # Dequeue a vertex from the queue and print it
            s = queue.pop(0)
            print(s, end=" ")

            # Get all adjacent vertices of the dequeued vertex
            # If an adjacent vertex has not been visited, mark it and enqueue it
            for i in self.graph[s]:
                if i not in visited:
                    queue.append(i)
                    visited.append(i)

# Driver code
if __name__ == "__main__":
    # Create a graph
    g = Graph()
    g.add_edge(0, 1)
    g.add_edge(0, 2)
    g.add_edge(1, 2)
    g.add_edge(2, 0)
    g.add_edge(2, 3)
    g.add_edge(3, 3)

    print("Following is Depth First Traversal (starting from vertex 2):")
    g.dfs(2)

    print("\nFollowing is Breadth First Traversal (starting from vertex 2):")
    g.bfs(2)
```

```python
from collections import deque

# Function to find the solution to the water jug problem
def solve_jug_problem(a, b, target):
    # Dictionary to mark visited states
    visited = {}
    is_solvable = False
    path = []

    # Initialize queue with the starting state (both jugs empty)
    queue = deque()
    queue.append((0, 0))

    while queue:
        # Current state
        u = queue.popleft()

        # Skip if already visited
        if u in visited:
            continue

        # Mark the current state as visited
        visited[u] = True
        path.append([u[0], u[1]])

        # Check if we have reached the target
        if u[0] == target or u[1] == target:
            is_solvable = True
            if u[0] == target and u[1] != 0:
                path.append([u[0], 0])
            elif u[1] == target and u[0] != 0:
                path.append([0, u[1]])
            break

        # Generate possible next states
        # Fill Jug1
        queue.append((a, u[1]))
        # Fill Jug2
        queue.append((u[0], b))
        # Empty Jug1
        queue.append((0, u[1]))
        # Empty Jug2
        queue.append((u[0], 0))
        # Pour Jug1 to Jug2
        pour_to_jug2 = min(u[0], b - u[1])
        queue.append((u[0] - pour_to_jug2, u[1] + pour_to_jug2))
        # Pour Jug2 to Jug1
        pour_to_jug1 = min(u[1], a - u[0])
        queue.append((u[0] + pour_to_jug1, u[1] - pour_to_jug1))

    # Print the solution path
    if is_solvable:
        print("Path from initial state to solution state:")
        for state in path:
            print(f"({state[0]}, {state[1]})")
    else:
        print("Solution not possible")

# Driver code
if __name__ == "__main__":
    jug1, jug2, target = 4, 3, 2
    print("Solving the Water Jug Problem:")
    solve_jug_problem(jug1, jug2, target)
```
